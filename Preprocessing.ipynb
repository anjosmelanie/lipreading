```python
import cv2
vidcap = cv2.VideoCapture('./drive/MyDrive/AT-20180422.mp4')
fps = vidcap.get(cv2.CAP_PROP_FPS)
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite(f"images/frame{count:05d}_{int(count/fps*1000)}.jpg", image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
